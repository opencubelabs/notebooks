```python
import paho.mqtt.client as mqtt
from pymongo import MongoClient
from bson.json_util import dumps
import time

client = MongoClient()
db = client.test_db

def on_message(client, userdata, message):
    print("message received " ,str(message.payload.decode("utf-8")))
    print("message topic=",message.topic)
    print("message qos=",message.qos)
    print("message retain flag=",message.retain)
    
    cur = db.data_col.insert_one({'msg': str(message.payload.decode("utf-8"))})
    data = json.loads(dumps(cur))
    print(data)


broker_address="iot.eclipse.org"

print("New instance created")
client = mqtt.Client("SU_1")
client.on_message=on_message

print("Connecting to broker...")
client.connect(broker_address)
client.loop_start()

print("Subscribing to topic","su_test123")
client.subscribe("su_test123")

print("Publishing msg to topic","su_test123")
client.publish("su_test123","testing")

time.sleep(4)
client.loop_stop()
```
